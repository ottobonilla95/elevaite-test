```python
import os
import json
```

```python
# Load inputs
input_file = os.environ.get("INPUT_JSON", "input_data.json")

try:
    with open(input_file, "r") as f:
        inputs = json.load(f)
except FileNotFoundError:
    inputs = {}
```

```python
# Simulate a processing step
task1_output = inputs.get("task1_output", "default_value") + "_processed"
task2_output = task1_output + "_final"
```

```python
# Print results
print(f"Processed Task 1 Output: {task1_output}")
print(f"Final Task 2 Output: {task2_output}")
```

```python
# Save outputs to a file (this file path would be referenced in the pipeline)
output_file = "output_data.json"
outputs = {"task1_output": task1_output, "task2_output": task2_output}

with open(output_file, "w") as f:
    json.dump(outputs, f)

print(f"Outputs saved to {output_file}")
```
